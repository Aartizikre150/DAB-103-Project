**Q2: Are some medical procedures common or frequent in some provinces compared to other provinces?**

```python
# Filter the data
filtered_df = data[(data['Region'].isna()) & 
                 (data['year'] == 2022) & 
                 (data['unitOfMeasurement'] == 'Number of cases')].copy()

# remove rows with NaN values for the "Indicator result" column
filtered_df = filtered_df.dropna(subset=['indicatorResult'])

# Exclude "Canada" from the list of Province
filtered_df = filtered_df[~filtered_df['Province'].isin(['Canada'])]

# Exclude "CT scan" and "MRI scan" from the list of surgeries (indicators)
filtered_df = filtered_df[~filtered_df['Indicator'].isin(['CT Scan', 'MRI Scan'])]

# Change the value 'Hip Fracture Repair/Emergency and Inpatient' to 'Hip Fracture Repair' in the Indicator column
filtered_df['Indicator'] = filtered_df['Indicator'].replace('Hip Fracture Repair/Emergency and Inpatient', 'Hip Fracture Repair')

# Get a list of all the provinces/territories in the filtered_df
provinces = filtered_df['Province'].unique()

# Define the number of rows and columns for the subplots
num_rows = (len(provinces) + 1) // 2
num_cols = 2

# Create the subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(12, 6*num_rows))

# Iterate over the provinces and plot the bar chart for each one
for i, province in enumerate(provinces):
    # Filter the filtered_df to only include the current province
    province_filtered_df = filtered_df[filtered_df['Province'] == province]

    # group the data by the indicator category and sum the indicator result
    province_filtered_df = province_filtered_df.groupby(['Indicator', 'Metric']).agg({'indicatorResult': 'sum'}).reset_index()

    # sort the data by indicator result in descending order
    province_filtered_df = province_filtered_df.sort_values(by='indicatorResult', ascending=False)

    # Get the names of all the surgeries for the current province
    surgeries = province_filtered_df['Indicator'].tolist()

    # Get the number of surgeries for each type for the current province
    surgery_counts = province_filtered_df['indicatorResult'].tolist()

    # Only plot the bar chart if there is data available
    if surgeries and surgery_counts:
        # Calculate the row and column index for the subplot
        row = i // num_cols
        col = i % num_cols

        # Select the axis for the current subplot
        if num_rows == 1:
            ax = axs[col]
        else:
            ax = axs[row, col]

        # Create a bar chart showing the number of surgeries for each type
        bars = ax.barh(surgeries, surgery_counts, color = '#5ea5c5')

        # Set the title and labels for the chart
        ax.set_title(f'Surgeries in {province}')
        ax.set_ylabel('Surgery Type')
        ax.set_xlabel('Number of Cases')

        # Add data labels to the bars without overlapping data
        for rect in bars:
            width = rect.get_width()
            if width > 0:
                ax.text(width + 50, rect.get_y() + rect.get_height() / 2, f'{int(width)}', ha='left', va='center')

        # Adjust the plot margins to prevent labels from going outside the plot area
        ax.margins(x=0.18)

plt.tight_layout()

# Show the chart
plt.show()
```
